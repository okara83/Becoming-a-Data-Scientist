# Runge Kutta (RK) fourth-order method for solving ordinary differential equations

RK4 can be used to find solutions for differential equations such as

$\frac{dy}{dx} = x^2 - \frac{1}{y} $

with initial condition $y = y_0$ at $x = x_0$. The point at which we wish to know the curve between the initial values and those of $x_n$ and $y_n$. And where n represent the number of step taken between $x_0$ to $x_n$

# Function to be solved

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y)->float: # function to be solved
    '''
    this corresponds to the differential
    equation dy/dx = x^2 - 1/y
    '''
    return x**2 - 1/y
# or
# f = lambda x: x+y
```

# RK-4 Method

```python
def rk4(f,x0,y0,xn,n):
    '''
       f = function to evalutate
       (x0,y0) = initial conditions
       x0 = initial x coordinate
       y0 = initial y coordinate
       xn = evaluation point to find yn
       n = number of steps to estimate xn
    '''
    del_x=(xn-x0)/n  # Calculating step size
    coordinates=[]
    for i in range(n):
        k1 = del_x*(f(x0, y0))
        k2 = del_x*(f((x0+del_x/2),(y0+k1/2)))
        k3 = del_x*(f((x0+del_x/2),(y0+k2/2)))
        k4 = del_x*(f((x0+del_x),(y0+k3)))
        k = (k1+2*k2+2*k3+k4)/6
        yn = y0+k
        y0 = yn
        x0 = x0+del_x
        coordinates.append([x0,y0])
    return coordinates
```

```python
rk4(f,1,2,4,10)
```

```python
rk4_array = rk4(f,1,2,4,10)
xx=[]
yy=[]
for i in rk4_array:
    xx.append(i[0])
    yy.append(i[1])
```

```python
plt.plot(xx,yy)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
