# Uniform Distribution

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
(a, b) = (0, 30)
```

```python
unifDist = stats.uniform(a, b)
```

```python
unifDist.args
```

```python
unifDist.pdf(10)
```

```python
unifDist.pdf(15)
```

```python
unifDist.pdf(30)
```

```python
unifDist.pdf(31) #Out of range --> 0
```

```python
unifDist.cdf(15) - unifDist.cdf(10)
```

```python
a = unifDist.rvs(1000)
b = []
for i in a:
    b.append(unifDist.pdf(i))
```

```python
plt.plot(a,b)
plt.xlim(-10, 40)
plt.show()
```

```python
mean, var = unifDist.stats()
```

```python

# printing mean and variance 
print("mean = " + str(mean)) 
print("variance = " + str(var))
print("median = " + str(unifDist.median())) 
print("standard deviation = " + str(unifDist.std()))
```

# Normal Distribution

```python
(mu, sigma) = (0, 1)
```

```python
normDist = stats.norm(mu, sigma)
```

```python
normDist.args
```

```python
1- normDist.cdf(2)
```

```python
normDist.cdf(2) - normDist.cdf(-2)
```

```python
normDist.cdf(1.73) - normDist.cdf(0)
```

```python
1-.0968
```

```python
stats.norm(300, 50).cdf(365)
```

```python
x = np.linspace(-5,5,1000)

y = normDist.pdf(x)

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
mean, var, skew, kurt = normDist.stats(moments='mvsk')
```

```python
# printing mean and variance 
print("mean = "+str(mean)) 
print("variance = "+str(var))
print("median = "+str(normDist.median())) 
print("standard deviation = "+str(normDist.std()))
print("skewness = "+ str(skew))
print("kurtosis = "+ str(kurt))
```

# t Distribution

```python
stats.t.cdf(-0.7745966, df=14)
```

```python
stats.t.cdf(0, df=14)
```

```python
tDist = stats.t(df=15)

x = np.linspace(-5, 5, 100)

y = tDist.pdf(x)
```

```python
plt.plot(x, y)

plt.grid()

plt.show()
```
