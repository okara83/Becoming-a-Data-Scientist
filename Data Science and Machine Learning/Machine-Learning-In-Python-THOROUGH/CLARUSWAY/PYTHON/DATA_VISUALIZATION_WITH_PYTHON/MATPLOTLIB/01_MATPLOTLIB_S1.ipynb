```python
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]

salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age, salary);
```

```python
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

```python
plt.plot(age, salary, label="salary")
plt.plot(age, salary_2, label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")
plt.legend();
```

```python
plt.subplot(2,1,1)
plt.plot(age, salary, "r")
plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.tight_layout();
```

```python
plt.subplot(1,2,1)
plt.plot(age, salary, "r")
plt.subplot(1,2,2)
plt.plot(age, salary_2, "b")
plt.tight_layout();
```

```python
plt.subplot(2,2,1)
plt.plot(age, salary, "r")
plt.subplot(2,2,2)
plt.plot(age, salary_2, "b")
plt.subplot(2,2,3)
plt.plot(age, salary, "g")
plt.subplot(2,2,4)
plt.plot(age, salary_2, "y")
plt.tight_layout();
```

```python
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, 0.8, 0.8])
```

```python
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, 0.8, 0.8])

ax.plot(age,salary, "r");
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age");
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0, 0.0, 0.8, 0.8])

ax2 = fig.add_axes([0.9, 0.3, 0.4, 0.4])


ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")

ax2.plot(age,salary_2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");
```

```python
fig, ax = plt.subplots(2, 1)
a=ax[0].plot(age, salary,"r")
ax[0].set_title("First Plot")

b=ax[1].plot(age, salary_2, "b")
ax[1].set_title("Second Plot")

plt.tight_layout();
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(age,salary, "r")
ax[0,0].set_title("salary by age")
ax[0,1].plot(age, salary_2)
ax[0,1].set_title("salary_2 by age")
ax[1,0].plot(age,salary, "g")
ax[1,0].set_title("salary by age")
ax[1,1].plot(age, salary_2, "y")
ax[1,1].set_title("salary_2 by age")
plt.tight_layout()
```
